```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://www.basketball-reference.com/leagues/NBA_2024_per_game.html')
```

```python
soup = BeautifulSoup(page.content, 'lxml')
```

```python
table = soup.find('table')
```

```python
while table.find('tr', class_ = 'thead') is not None :
    table.find('tr', class_ = 'thead').decompose()
```

```python
pd.read_html(str(table))[0]
```
